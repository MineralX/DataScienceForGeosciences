# K-means clustering on the nutrition dataset

```
from sklearn.cluster import KMeans
# define cluster number and perform k-means clustering
cluster_num = 3
kmeans = KMeans(n_clusters=cluster_num, random_state=0).fit(nutrition_data[nutrition_data.columns[:9]])
## get clustering labels
nutrition_data['kmeans_label'] = kmeans.labels_
nutrition_data['kmeans_label'] = nutrition_data['kmeans_label'].astype("category")
## get cluster centers
cluster_center = kmeans.cluster_centers_
```
